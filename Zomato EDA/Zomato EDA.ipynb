```python
df_country = pd.read_excel('Country-Code.xlsx')
```

```python
df_country
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
final_df.head()
```

```python
final_df.dtypes
```

```python
country_name=final_df.Country.value_counts().index
```

```python
country_values = final_df.Country.value_counts().values
```

```python
#plt pie chart- Top 3 countries that uses zomato
plt.pie(country_values[:3],labels=country_name[:3],autopct='%1.2f%%');
```

**observation : zomato maximum records and transactions are from in India followed by USA and United Kingdom**

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0 :'Rating count'})
ratings
```

#Observation 
1. when rating is between 4.5 to 4.9 food was excellent
2. when ratings are between 4.0 to 4.4 ratings are very good
3. when rating is between 3.5 to 3.9 ratings are good
4. when rating is between 2.5 to 3.4 ratings are average
5. when rating is between 2.0 to 2.4 ratings are poor


```python
plt.figure(figsize=(12,6))
sns.barplot(x='Aggregate rating', y= 'Rating count',data=ratings);
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Aggregate rating', y= 'Rating count',data=ratings,hue='Rating color',palette=['blue','red','orange','yellow','green','green']);
```

observation: 
1. not rated count is very high
2. max number of ratings are between 2.5 to 3.5 

```python
## count plot
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
